```python
# Exercise 4
def func4(a, b): #function defining
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a: # a loop for every x in a
        if binary_search(b, x) >= 0: # condition if binary search greater than zero then only proceed to next step
            intersect = intersect + 1 #after above operation there will be an increment in intersect
    return intersect # then intersect will be return as a partial result

def binary_search(arr, target): #defining another func name binary_search
    low, high = 0, len(arr) - 1 #declaring variables and initilize them with 0
    while low <= high: #condition if low less than or equals to high then
        mid = (low + high) // 2  # a third variable to find the mid value in binary searching array
        if arr[mid] == target: # condition if the mid value is equal to target then mid will become target
            return mid # and it will return mid
        elif arr[mid] < target: # otherwise if mid is less than target
            low = mid + 1 # move the index of mid to one increment of low
        else:
            high = mid - 1 #else move the index of mid to one increment of high
    return -1
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  total = 0
  for num in arr:
    total += num
    return total
```

```python
#call the function
#arr = [12, 3, 4, 15,45,78,450]
#print(sum_arr)
arr = [12,3,4,15,45,78,450]
result = sum_arr(arr)
print(result)
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  if len(arr) == 0:
    return None, None
  maximum = arr[0]
  minimum = arr[0]
  for i in range(1, len(arr)):
    if arr[i] >maximum:
      maximum = arr[i]
    elif arr[i] < minimum:
      minimum = arr[i]
  return maximum, minimum
arr = [12, 3, 4, 15, 45, 78, 450]
max_val, min_val = max_min(arr)
print("Maximum element:", max_val)
print("Minimum element:", min_val)
```

```python
# call sum_arr to find sum of elements in given array for arr = [12, 1234, 45, 67, 1]
def sum_arr(arr):
    total = 0
    for num in arr:
        total += num
    return total
arr = [12, 1234, 45, 67, 1]
result = sum_arr(arr)
print("The sum of the array is:", result)
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0
    for num in arr:
      if num % 2 == 0:
            even_count += 1
      else:
            odd_count += 1
    return even_count, odd_count
arr = [12, 1234, 45, 67, 1]
evens, odds = Counting_EvenOdd(arr)
print(f"Even numbers: {evens}")
print(f"Odd numbers: {odds}")
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
arr = [1, 3, 5, 7, 9]
target = 7
result = linear_Search(arr, target)
target_missing = 4
result_missing = linear_Search(arr, target_missing)
print(f"Index of {target}: {result}")
print(f"Index of {target_missing}: {result_missing}")
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
arr = [1, 3, 5, 7, 9, 14, 15, 16, 45]
target = 14
result = Binary_Search(arr, target)
print(f"Index of {target} in sorted array: {result}")
```
